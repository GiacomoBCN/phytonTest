## Plot a histogram showcasing the age distribution of users.

```python
df['gender'].unique()
```

```python

filtered_df = df[df['gender'].isin(['M', 'F'])]
filtered_df.groupby('gender')['age'].hist(alpha=0.6)
plt.show()
```

```python

# Filter only Male and Female
filtered_df = df[df['gender'].isin(['M', 'F'])]

# Define custom colors
colors = {'M': 'skyblue', 'F': 'pink'}

# Plot histograms per gender
for gender, group in filtered_df.groupby('gender'):
    group['age'].hist(alpha=0.6, label=gender, color=colors[gender])

# Add legend and show plot
plt.legend(title='Gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution by Gender')
plt.show()
```
