### Feature distributions

Check the distributions of numerical features (histogram or boxplot) and categorical features (bar plot).

```python
df.hist(figsize=(12, 10))
plt.tight_layout()
plt.show()
```

```python
df.plot(kind='box', figsize=(12, 6))
plt.title('Boxplot of All Numeric Features')
plt.show()

for col in df.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(5, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
print ('Histograma + Boxplot por cada columna')

sns.set_theme(style="whitegrid")

# Iterar sobre columnas numéricas
sns.set(style="whitegrid")

for col in df.select_dtypes(include=['number']).columns:
    # Limpiar datos: coercionar, quitar inf y NaN
    s = pd.to_numeric(df[col], errors='coerce')
    s = s.replace([np.inf, -np.inf], np.nan).dropna()

    # Si no quedan datos, saltar columna
    if s.empty:
        print(f"Saltando '{col}' porque no tiene datos válidos para boxplot/hist.")
        continue

    # Figura compacta: hist arriba + box abajo
    fig, axes = plt.subplots(2, 1, figsize=(4.5, 4.5))

    # Histograma (sin especificar colores si no quieres)
    sns.histplot(s, kde=True, bins=30, ax=axes[0])
    axes[0].set_title(f'Histogram of {col}', fontsize=10)

    # Boxplot horizontal
    sns.boxplot(x=s, ax=axes[1])
    axes[1].set_title(f'Boxplot of {col}', fontsize=10)

    plt.tight_layout()
    plt.show()
```

### Feature correlation

Display the correlation between pairs of features. This is normally done with the correlation matrix.

Also, display in the same plot or a different one the correlation of features with the target feature.

```python
df.corr(numeric_only= True)
```

```python
#imprimimos la matrix de correlacion con seborn

mat_corr = df.corr(numeric_only= True)
plt.figure(figsize=(15,15))
sns.heatmap(mat_corr,
            annot= True,
            fmt= ".2f",
            cmap='coolwarm',
            linewidths=0.5)
plt.title('Matriz de correlacion entre variables')
plt.tight_layout()
plt.show
```

### Pairplot

Visualize the relationship of different features with a pairplot.

```python
sns.pairplot(
    mat_corr,
    diag_kind='kde',
    corner=True)
plt.suptitle('relaciones entre variables mumericas')
plt.show()
```

```python
plt.figure(figsize=(15, 15))

# crea el pairplot
g = sns.pairplot(
    data=df,
    vars=['SalePrice', 'YearBuilt', 'GrLivArea', 'TotRmsAbvGrd']
)

# añade título general
g.fig.suptitle('Gráfico de las variables correlacionadas', y=1.02)

plt.show()

 
```

## Verify Data Quality

- Is the data correct? Does any column contain important errors that should be corrected?
- What about missing values?
    - How are they represented (e.g., empty string, NaN, -1, etc.)?
    - Where do they occur?
    - How common are they?
- Do not change the data, just inspect where these problems occur.

```python
# Basic info: data types + non-null counts
print(df.info())

# Number of missing values per column
print("\nMissing values per column:\n")
print(df.isna().sum().sort_values(ascending=False))

# Percentage of missing values
print("\nPercentage of missing values per column:\n")
print((df.isna().mean() * 100).round(2))
```

```python
# Check unusual representations of missing values in object columns
for col in df.select_dtypes(include='object').columns:
    uniques = df[col].unique()
    print(f"{col}: {len(uniques)} unique values -> {uniques[:10]}")
```

```python
# Find columns with impossible or suspicious values
print(df.describe(include='all').T)

# Example: check if any year > current year
print(df[df['YearBuilt'] > 2025])

# Example: negative or zero lot areas
print(df[df['LotArea'] <= 0])
```

```python
# Check duplicates
print("Number of duplicated rows:", df.duplicated().sum())

# If you have an ID column:
print("Duplicate IDs:", df['Id'].duplicated().sum())
```
