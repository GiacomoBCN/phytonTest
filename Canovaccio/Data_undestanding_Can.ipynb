#### 4.3 Distribuzioni numeriche

```python
# Histogrammi per TUTTE le colonne numeriche insieme
df.hist(figsize=(12, 10))
plt.tight_layout()
plt.show()
```

```python
# Histogram + Boxplot VERTICALE per OGNI colonna numerica (uno per uno)
print('Histograma + Boxplot por cada columna')

sns.set_theme(style="whitegrid")

# Iterar sobre columnas numéricas
for col in df.select_dtypes(include=['number']).columns:
    # Limpiar datos: coercionar, quitar inf y NaN
    s = pd.to_numeric(df[col], errors='coerce')
    s = s.replace([np.inf, -np.inf], np.nan).dropna()

    # Si no quedan datos, saltar columna
    if s.empty:
        print(f"Saltando '{col}' porque no tiene datos válidos para boxplot/hist.")
        continue

    # Figura: hist a SINISTRA + box a DESTRA (verticale)
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))

    # Histograma (a sinistra)
    sns.histplot(s, kde=True, bins=30, ax=axes[0])
    axes[0].set_title(f'Histogram of {col}', fontsize=10)

    # Boxplot VERTICALE (a destra)
    sns.boxplot(y=s, ax=axes[1])
    axes[1].set_title(f'Boxplot of {col}', fontsize=10)

    plt.tight_layout()
    plt.show()

    # subplots(1, 2) → 1 riga, 2 colonne (fianco a fianco)
    # sns.boxplot(y=s) → boxplot verticale (prima era x=s orizzontale)
    # figsize=(10, 4) → più largo per i 2 grafici affiancati
```

#### 5.4 Controllare valori impossibili Es: anni > 2025, valori negativi stran

```python
# Tabella con min e max per TUTTE le colonne numeriche
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

summary = pd.DataFrame({
    'Min': df[numeric_cols].min(),
    'Max': df[numeric_cols].max()
})
print(summary)
```

```python
# Controllare anni futuri (MODIFICA anno_limite)
anno_limite = 2025  # ← CAMBIA QUI
year_columns = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']  # ← CAMBIA QUI

year_check = []
for col in year_columns:
    if col in df.columns:
        count_future = (df[col] > anno_limite).sum()
        if count_future > 0:
            year_check.append({
                'Colonna': col,
                f'Valori > {anno_limite}': count_future
            })

# Mostra risultati
if len(year_check) > 0:
    print(f"ATTENZIONE: Anni > {anno_limite} trovati!")
    print(pd.DataFrame(year_check))
else:
    print(f"✓ Nessun anno > {anno_limite}")
```

```python
# Tabella con valori negativi per ogni colonna numerica
negative_check = []
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    negativi = (df[col] < 0).sum()
    if negativi > 0:
        negative_check.append({
            'Colonna': col,
            'N. Negativi': negativi,
            'Min': df[col].min()
        })

# Mostra risultati
if len(negative_check) > 0:
    print("ATTENZIONE: Valori negativi trovati!")
    print(pd.DataFrame(negative_check))
else:
    print("✓ Nessun valore negativo trovato")
```

```python
# Controllare valori <= 0 dove NON dovrebbero esserci
suspicious_cols = ['LotArea', 'SalePrice']  # ← CAMBIA QUI con le tue colonne

suspicious_check = []
for col in suspicious_cols:
    if col in df.columns:
        count_zero_neg = (df[col] <= 0).sum()
        if count_zero_neg > 0:
            suspicious_check.append({
                'Colonna': col,
                'Valori <= 0': count_zero_neg
            })

# Mostra risultati
if len(suspicious_check) > 0:
    print("ATTENZIONE: Valori <= 0 trovati!")
    print(pd.DataFrame(suspicious_check))
else:
    print("✓ Nessun valore <= 0 nelle colonne controllate")
```

####   5.5 Controllare duplicatidf.duplicated().sum()

```python
# Controllare righe duplicate (tutte le colonne identiche)
n_duplicati = df.duplicated().sum()

if n_duplicati > 0:
    print(f"ATTENZIONE: {n_duplicati} righe duplicate trovate!")
    print("\nRighe duplicate:")
    print(df[df.duplicated(keep=False)].sort_values(by=list(df.columns)))
else:
    print("✓ Nessuna riga duplicata trovata")
```

```python
# Controllare duplicati in colonna ID (MODIFICA 'Id' con il nome della tua colonna)
id_column = 'Id'  # ← CAMBIA QUI

if id_column in df.columns:
    n_duplicati_id = df[id_column].duplicated().sum()
    
    if n_duplicati_id > 0:
        print(f"ATTENZIONE: {n_duplicati_id} ID duplicati trovati nella colonna '{id_column}'!")
        print("\nID duplicati:")
        duplicati = df[df[id_column].duplicated(keep=False)].sort_values(by=id_column)
        print(duplicati[[id_column]])
    else:
        print(f"✓ Nessun ID duplicato nella colonna '{id_column}'")
else:
    print(f"Colonna '{id_column}' non trovata nel dataset")
```

```python
# Controllare duplicati in più colonne specifiche (MODIFICA lista)
# Controllare case identiche (possibile errore)
# 2 case con stessa area, anno, metri quadri → sospetto!

columns_to_check = ['LotArea', 'YearBuilt', 'GrLivArea', 'GarageCars' ]  # ← CAMBIA QUI

if all(col in df.columns for col in columns_to_check):
    n_duplicati_subset = df.duplicated(subset=columns_to_check).sum()
    
    if n_duplicati_subset > 0:
        print(f"ATTENZIONE: {n_duplicati_subset} duplicati trovati nelle colonne {columns_to_check}!")
        print("\nRighe duplicate:")
        print(df[df.duplicated(subset=columns_to_check, keep=False)].sort_values(by=columns_to_check))
    else:
        print(f"✓ Nessun duplicato nelle colonne {columns_to_check}")
else:
    print(f"Alcune colonne non trovate: {columns_to_check}")
```

```python
# Controllare duplicati in più colonne specifiche 
# UTILE per trovare righe "sospette" anche con ID diversi
columns_to_check = ['MSZoning', 'LotArea', 'HouseStyle', 'YearBuilt']  # ← CAMBIA QUI

if all(col in df.columns for col in columns_to_check):
    n_duplicati_subset = df.duplicated(subset=columns_to_check, keep=False).sum()
    
    if n_duplicati_subset > 0:
        print(f"ATTENZIONE: {n_duplicati_subset} righe con stessi valori in {columns_to_check}!")
        print("\nRighe sospette (ordinate):")
        duplicati = df[df.duplicated(subset=columns_to_check, keep=False)].sort_values(by=columns_to_check)
        print(duplicati[columns_to_check + ['Id', 'SalePrice']].head(20))  # Mostra prime 20
    else:
        print(f"✓ Nessun duplicato nelle colonne {columns_to_check}")
else:
    print(f"Alcune colonne non trovate: {columns_to_check}")
```

#### 6.2 Bilanciamento classi (solo per CLASSIFICAZIONE) classiGrafico: df['target'].value_counts().plot(kind='bar')

```python
# MODIFICA 'target' con il nome della tua colonna target
target_column = 'SalePrice'  # ← CAMBIA QUI

# Controllare se il target è categorico (per classificazione)
if target_column in df.columns:
    if df[target_column].dtype == 'object' or df[target_column].nunique() < 20:
        print(f"Bilanciamento delle classi per '{target_column}':\n")
        
        # Tabella con conteggi e percentuali
        balance_table = pd.DataFrame({
            'Conteggio': df[target_column].value_counts(),
            'Percentuale': df[target_column].value_counts(normalize=True) * 100
        })
        print(balance_table)
        
        # Grafico a barre
        df[target_column].value_counts().plot(
            kind='bar', 
            color='steelblue', 
            edgecolor='black', 
            figsize=(8, 5)
        )
        plt.title(f'Bilanciamento classi: {target_column}')
        plt.xlabel(target_column)
        plt.ylabel('Frequenza')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
        
        # Messaggio se sbilanciato
        max_class = df[target_column].value_counts().max()
        min_class = df[target_column].value_counts().min()
        ratio = max_class / min_class
        
        if ratio > 3:
            print(f"\n⚠️ DATASET SBILANCIATO! Ratio: {ratio:.2f}:1")
        else:
            print(f"\n✓ Dataset abbastanza bilanciato. Ratio: {ratio:.2f}:1")
    else:
        print(f"'{target_column}' è numerico → usa analisi per REGRESSIONE")
else:
    print(f"✗ Colonna '{target_column}' non trovata")
```

#### 6.3 - Relazione target vs features (mejor no hacerlo por leakeg)


✅ Tre blocchi:

Boxplot → target categorico vs features numeriche (CLASSIFICAZIONE)
Scatterplot → target numerico vs features numeriche (REGRESSIONE)
Crosstab + Stacked bar → target categorico vs features categoriche

Modifica:

target_column → nome colonna target
numeric_features → lista features numeriche
categorical_features → lista features categoriche

```python
# MODIFICA questi parametri
target_column = 'y'  # ← CAMBIA QUI con nome colonna target
numeric_features = ['age', 'balance', 'duration', 'campaign'] # ← CAMBIA QUI con tue features numeriche

# Per CLASSIFICAZIONE: Boxplot per ogni feature numerica vs target categorico
if target_column in df.columns:
    if df[target_column].dtype == 'object' or df[target_column].nunique() < 20:
        print(f"Relazione tra '{target_column}' e features numeriche:\n")
        
        for feature in numeric_features:
            if feature in df.columns:
                plt.figure(figsize=(8, 5))
                sns.boxplot(x=target_column, y=feature, data=df)
                plt.title(f'{feature} vs {target_column}')
                plt.xticks(rotation=45)
                plt.tight_layout()
                plt.show()
            else:
                print(f"✗ Feature '{feature}' non trovata")
    else:
        print(f"'{target_column}' è numerico → analisi regressione")
```

```python
# Per REGRESSIONE: Scatterplot per ogni feature numerica vs target numerico
if target_column in df.columns:
    if df[target_column].dtype in ['int64', 'float64']:
        print(f"Relazione tra '{target_column}' e features numeriche:\n")
        
        for feature in numeric_features:
            if feature in df.columns:
                plt.figure(figsize=(8, 5))
                plt.scatter(df[feature], df[target_column], alpha=0.5)
                plt.xlabel(feature)
                plt.ylabel(target_column)
                plt.title(f'{feature} vs {target_column}')
                plt.tight_layout()
                plt.show()
            else:
                print(f"✗ Feature '{feature}' non trovata")
```

```python
# Relazione target vs features CATEGORICHE (per classificazione)
categorical_features = ['job', 'marital', 'education']  # ← CAMBIA QUI

if target_column in df.columns:
    if df[target_column].dtype == 'object' or df[target_column].nunique() < 20:
        for feature in categorical_features:
            if feature in df.columns:
                # Crosstab
                crosstab = pd.crosstab(df[feature], df[target_column], normalize='index') * 100
                print(f"\n{feature} vs {target_column} (%):")
                print(crosstab)
                
                # Grafico
                crosstab.plot(kind='bar', stacked=True, figsize=(10, 5))
                plt.title(f'{feature} vs {target_column}')
                plt.xlabel(feature)
                plt.ylabel('Percentuale')
                plt.xticks(rotation=45)
                plt.legend(title=target_column)
                plt.tight_layout()
                plt.show()
            else:
                print(f"✗ Feature '{feature}' non trovata")
```
