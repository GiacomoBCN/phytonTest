#### Categorical features

```python
import math

cats = df.select_dtypes(include=['object', 'category'])
n = len(cats.columns)
cols_per_row = 4

rows = math.ceil(n / cols_per_row)
fig, axes = plt.subplots(rows, cols_per_row, figsize=(5*cols_per_row, 4*rows))
axes = axes.flatten()

for i, col in enumerate(cats.columns):
    df[col].value_counts().plot(kind='bar', ax=axes[i], title=col)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Hide any unused subplots
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

Some of the values have a very small amount of categories.

After checking them (code below) the conclusions are:

- `MSZoning` has just two commercial... The rest are OKeish.
- `LandSlope` has just a few examples of Sev slope. It probably has no effect but we can keep it.
- `HouseStyle` only 2 2.5Fin and 5 2.5Unf. As SLvl and SFoyer are very similar to finished 1.5Fin, we  can include them there.
- `RoofStyle` seems useful but has only 3 of Flat and Gambrel, which is very low.
- `Foundation` seems ok although it has the Stone with just 4 instances.
- `Street` has mostly Pave but a few Grvl as well. As having Grvl could be an indicator of lower price we can keep it.
- `Utilities` only has one example of NoSeWa --> useless column.
- `Heating` has mostly GasA values. The second most frequent is GasW with 10... which is also gas. We can probably drop this feature.
- `PavedDrive` has mostly Y values, but some of the others as well. Leave it as it is for now.
- `Electrical` has FuseF and FuseP with very few examples. We can proably create a boolean feature that indicates `SBrkr` or `Fuse` (with A, F and P together), although there is a big difference between FuseA and FuseF (limited power) and FuseP (fire risk). Maybe we  can create a category FuseFP that combines both bad fuses. We'll check the NaNs later.
- `MiscFeature` has only 1 example of TenC and Othr, these categories are not useful. We can create a new column `HasShed` instead of this feature.

```python
to_check = [
    'MSZoning', 'LandSlope', 'HouseStyle', 'RoofStyle', 'Foundation',
    'Street', 'Utilities', 'Heating', 'PavedDrive', 'Electrical', 'MiscFeature']

for col in to_check:
    print(df[col].value_counts(dropna=False))
    print()
```

### Nulls

Conclusions after analysis below:

- `Electrical` null should probably be SBrkr.
- `GarageType` null means no garage.
- `GarageYrBlt` null means no garage.
- `GarageFinish` null means no garage.
- `MiscFeature` null means nothing to be added (data description). We can compute a new feature `HasShed` and drop this one.
- `SalePrice` nulls correspond to leaderboard prices we don't know. They're OK.
- `Electtrical` is all nulls. It is not even defined in the data description. It is probably an erroneous feature we can remove.

```python
# Null proportion
prop_nans = df.isna().sum(axis=0) / len(df)
prop_nans[prop_nans > 0]
```

It is suspicious that all garage features have the same proportion of nulls.

As per the code below, all three work in conjunction, when one is null, so are the rest.

As in the description it is specified that a null `GarageFinish` means there is no garage, we know that NaN garage means there is no garage.

```python
# Indices of null GarageType values
set(df[df['GarageType'].isna()].index) == set(df[df['GarageYrBlt'].isna()].index) == set(df[df['GarageFinish'].isna()].index)
```
